```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


#  Graph Plotter Class
class GraphPlot():
    def __init__(self):
        self.function = input("Type a function(x): ")
        self.x_values = np.array([])
        self.y_values = np.array([])
        self.function_dataframe = pd.DataFrame() # Dataframe to store the values (x,y)
    # Function to return the typed function values
    def create_function(self, i):
        result = eval(self.function, {'x': i})
        return result
    # Function to create range values from (-20 to 20)
    def create_range_values(self):
        for i in np.arange(-20,21,0.5):
            self.x_values = np.append(self.x_values,i)
            self.y_values = np.append(self.y_values, self.create_function(i))
    # Function to store the created range values into a dataframe
    def create_dataframe(self):
        self.function_dataframe["X Values"] = self.x_values
        self.function_dataframe["Y Values"] = self.y_values
    # Function to plot the graph
    def create_graph(self):
        plt.style.use("seaborn-whitegrid")
        fig, ax = plt.subplots(figsize=(10,6))
        ax.plot(self.function_dataframe["X Values"], self.function_dataframe["Y Values"], color="black")
        ax.set(title="Graph -> " + self.function, xlabel="X Values", ylabel="Y Values")
        ax.axhline(0)
        ax.axvline(0);


plotter = GraphPlot()
plotter.create_range_values()
plotter.create_dataframe()
plotter.create_graph()
```
